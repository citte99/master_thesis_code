```python

import scipy
#now fit a curve
def relu_fit_curve(M_max, R_0, a, log_m_0):
    m_0=10**log_m_0
    M = np.asarray(M_max, dtype=float)        
    R = np.full_like(M, R_0, dtype=float)
    mask = M > m_0
    R[mask] = R_0 + a * np.log(M[mask] / m_0)

    # cast back to plain float if the input was scalar
    if np.isscalar(M_max):
        return float(R)
    return R



results, covariance=scipy.optimize.curve_fit(relu_fit_curve, mass_max_array, R, bounds=([-np.inf, -np.inf, 5], [+np.inf, +np.inf, +np.inf]))

plt.plot(mass_max_array, R)
plt.plot(mass_max_array, relu_fit_curve(mass_max_array, *results))
#plt.axline((0.0, 0.5), slope=0.0)
plt.xscale("log")
#plt.ylim(-1, 1.20)
plt.show()

print(f"R_0, a, log_m_0={results}")

print(mass_max_array)
print(relu_fit_curve(mass_max_array, *results))
```

```python
from scipy.special import erf     # SciPys erf works on arrays
def R_from_s(s):

    s = np.asarray(s, dtype=float)
    t = erf(s / np.sqrt(2.0))          # t  (1, 1)
    return np.log((1.0 + t) / (1.0 - t))
```

```python
R_lim=R_from_s(1.0)
print(R_lim)
```

```python
def get_relu_intersect(fit_results, sigma_detect):

    R0, a, log_m0 = fit_results
    m0 = 10.0**log_m0                      # convert back to linear mass
    R_th = R_from_s(sigma_detect)          # log-odds threshold

    R_th = np.asarray(R_th, dtype=float)   # allows vector input

    # Case 1: threshold below (or equal) to plateau value    first hit at m0
    mask_plateau = R_th <= R0
    M_star = np.where(mask_plateau, m0, m0 * np.exp((R_th - R0) / a))

    # Return scalar if the input was scalar
    return float(M_star) if np.isscalar(sigma_detect) else M_star    
```

```python
system_dict_base=system_dict


dataset_sm=SingleTelescopeNoiseDataset(
        catalog_dict=catalog_dict,
        grid_width_arcsec=6.0,
        grid_pixel_side=100,
        broadcasting=True,
        sky_level=0.01,
        kernel_size=5,
        kernel_sigma= 1.0,
        gain=1000.0
)



num_masses=20


from shared_utils import _grid_lens

nx = ny = 40                        # grid resolution
grid_tensor = _grid_lens(4.0, nx, device="cpu")   # (ny, nx, 2) in *radians*

grid_np = grid_tensor.numpy()        # torch  NumPy
pixels  = grid_np.reshape(-1, 2)     # (nx*ny, 2)  row-major order
# ------------------------------------------------------------------

# container that will hold the output map
M_detect_grid = np.full((ny, nx), np.nan, dtype=float)

from tqdm import tqdm


for idx, pix_coord in enumerate(tqdm(pixels)):
    mass_max_array=np.power(10, np.linspace(6, 11, num_masses))
    system_dict_list=[]
    for i, mass_max in enumerate(mass_max_array):
        system_dict=copy.deepcopy(system_dict_base)
        system_dict["system_index"]=i
        system_dict["lens_model"]["mass_components"][-1]["params"]["mass_max"]=mass_max
        system_dict["lens_model"]["mass_components"][-1]["params"]["r_max_kpc"]=get_r_max_from_M_max(mass_max)
        system_dict["lens_model"]["mass_components"][-1]["params"]["pos"]=pix_coord

        system_dict_list.append(system_dict)

    catalog_dict={"SL_systems":system_dict_list}
    
    dataset_sm.update_catalog_dict(catalog_dict)
    
    dataloader = custom_dataloader(dataset=dataset_sm,
                               batch_size=num_masses,
                               shuffle=False)
    myiterator=iter(dataloader)
    first_batch, _=next(myiterator)
    
    
    logits=classifier_stage.predict(first_batch)
    pr= torch.softmax(logits, dim=1)
    pr_pos=pr[:, 1].detach().numpy()
    pr_neg=pr[:, 0].detach().numpy()

    R=np.log10(pr_pos/pr_neg)
    results, covariance=scipy.optimize.curve_fit(relu_fit_curve, mass_max_array, R, bounds=([-np.inf, -np.inf, 7], [+np.inf, +np.inf, +np.inf]))
    M_intersect=get_relu_intersect(results, 1)
    
    # ------------ write result into the 2-D output grid ----------
    row, col = divmod(idx, nx)        # because pixels was ravelled row-major
    M_detect_grid[row, col] = M_intersect
    
    
    plt.plot(mass_max_array, R)
    plt.plot(mass_max_array, relu_fit_curve(mass_max_array, *results))
    #plt.axline((0.0, 0.5), slope=0.0)
    plt.xscale("log")
```
