```python
import numpy as np
import matplotlib.pyplot as plt

# Load the loss data with allow_pickle=True
loss_data = np.load("/raven/u/fcitterio/master_thesis_code/trained_classifiers/my_classifier/stages/ea22a6fb1ec0825bf0d5c18060323020/loss_data.npz", allow_pickle=True)

# Extract training and test losses
train_losses = loss_data['train_losses']
test_catalog_losses = loss_data.get('test_catalog_losses', {})

# Plot training losses
plt.figure(figsize=(8, 5))
plt.plot(train_losses, label="Training Loss", color="blue", marker="o")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss Over Epochs")
plt.legend()
plt.grid(True)
plt.show()

# If test catalog losses exist, visualize them
if test_catalog_losses:
    for catalog, losses in test_catalog_losses.item().items():
        plt.plot(losses, label=f"Test Loss ({catalog})", marker="x")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Test Loss Over Epochs")
    plt.legend()
    plt.grid(True)
    plt.show()
```
